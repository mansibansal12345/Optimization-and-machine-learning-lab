```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def contours(functions, titles, levels=20):
    # Create figure with 2x2 subplots
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))

    # Create grid points
    x1 = np.linspace(-4, 4, 200)
    x2 = np.linspace(-4, 4, 200)
    X1, X2 = np.meshgrid(x1, x2)

    # Plot each function
    for idx, (f, title) in enumerate(zip(functions, titles)):
        row = idx // 2
        col = idx % 2
        ax = axes[row, col]

        # Calculate Z values
        Z = f(X1, X2)

        # Create contour plot
        contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis')
        plt.colorbar(contour, ax=ax, label='f(x₁,x₂)')

        # Add labels and title
        ax.set_xlabel('x₁')
        ax.set_ylabel('x₂')
        ax.set_title(title, pad=10)
        ax.grid(True)
        ax.set_aspect('equal')

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Define the functions
def f1(x1, x2):
    return 100*x2**2 + 100*x1**4 - 200*x2*x1**2 + x1**2 - 2*x1 + 1

def f2(x1, x2):
    return x1**4 + x2**4 - 20*x1**3 - 20*x2**3 + 100*x1**2 + 100*x2**2

def f3(x1, x2):
    return -np.cos(x1)*np.cos(x2)*np.exp(-(x1-np.pi)**2 - (x2-np.pi)**2)

def f4(x1, x2):
    return 2*x1**2 - 1.05*x1**4 + (x1**6)/6 + x1*x2 + x2**2

# List of functions and their titles
functions = [f1, f2, f3, f4]
titles = ["f1(x)","f2(x)","f3(x)","f4(x)"]

# Create all plots
contours(functions, titles)
```
